4. Задание (в программе):
Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.


```python
import numpy as np
import pandas as pd
import random
%config InlineBackend.figure_format = 'svg'
from matplotlib import pyplot as plt
```

```python
x=np.linspace(-10,10,201)
```

```python
y1=np.cos(x)

y2=np.cos(2*x)

y3=np.cos(0.5*x)
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.title("Графики y1(x)")
plt.xlabel("Значения x")
plt.ylabel("Значения y1,y2")
plt.xlim(-8,8)
plt.ylim(-2.5,2.5)
plt.legend(labels=["y=cos(x)", "y=cos(2x)","y=cos(0.5x)"])
plt.show()
```
