3. Задание (в программе)
Напишите код на Python, реализующий построение графиков:
окружности,
эллипса,
гиперболы.


```python
x=[]
y1=[]
y2=[]
R=5
for i in range(-500,501):
    x_ = i/100
    x.append(x_)
    y1.append(math.sqrt(R**2-x_**2))
    y2.append(-math.sqrt(R**2-x_**2))

plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-7, 7)
plt.ylim(-7, 7)
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
```

```python
x=[]
y1=[]
y2=[]
a=4
b=0.4
for i in range(-a*100,a*100+1):
    x_ = i/100
    x.append(x_)
    y1.append(math.sqrt((1-(x_/a)**2)/b**2))
    y2.append(-math.sqrt((1-(x_/a)**2)/b**2))


plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
```

```python
x1=[]
x2=[]
y1=[]
y2=[]
y3=[]
y4=[]
a=3
b=0.5

for i in np.linspace(-50,-a,500):
    x_ = i
    x2.append(x_)
    y3.append(math.sqrt((-1+(x_/a)**2)/(b**2)))
    y4.append(-math.sqrt((-1+(x_/a)**2)/(b**2)))
for i in np.linspace(a,50,500):
    x_ = i
    x1.append(x_)
    y1.append(math.sqrt((-1+(x_/a)**2)/(b**2)))
    y2.append(-math.sqrt((-1+(x_/a)**2)/(b**2)))
    
plt.plot(x1,y1)
plt.plot(x1,y2)    
plt.plot(x2,y3)
plt.plot(x2,y4)
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-15, 15)
plt.ylim(-10, 10)
plt.gca().set_aspect('equal', adjustable='box')
plt.grid(True)
```

и прямая: 

Как узнать, принадлежит прямая плоскости или нет?

Ответ:должно выполнятся два условия

условие паралельности:  A1(x2-x1) + B1(y2-y1) + C1(z2-z1) =0

и условие принадлежности точки: A1x1 + B1y1 + C1z1 + D1 = 0

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig=figure()
ax=Axes3D(fig)
X=np.arange(-5,5,0.5)
Y=np.arange(-5,5,0.5)
X,Y=np.meshgrid(X,Y)
Z=3*X+2*Y
Z1=3*X+2*Y+26
ax.plot_wireframe(X,Y,Z)
ax.plot_surface(X,Y,Z1)
ax.scatter(0,0,0,'z',50,'red')
show()
```
