4. Задание (в программе)
Решите систему уравнений:
y = x2 – 1


exp(x) + x∙(1 – y) = 1




```python
from scipy.optimize import fsolve
x=np.linspace(-5,18,201)
f1=x**2-1
f2=1+(np.exp(x)-1)/x
      
def equations(p):
    x,y=p
    return (y-x**2+1, np.exp(x)+x*(1-y)-1)

plt.plot(x, f1)
plt.plot(x, f2)
plt.title("Графики f1(x), f2(x)")
plt.xlabel("Значения x")
plt.ylabel("Значения f(x)")
plt.xlim(-3,5)
plt.ylim(-2,20)
plt.legend(labels=["f1(x)", "f2(x)"])

plt.grid(True)
plt.show()
x1, y1=fsolve(equations,(-2,2))
print(x1,y1)
x2, y2=fsolve(equations,(2,3))
print(x2,y2)
x3, y3=fsolve(equations,(4,15))
print(x3,y3)
```

## Тут по моему пропушен корень 0,-1
поэтому я подставил Y во второе уравнение и получил выражение 
exp(x)+2x-x^3-1=0 и нашел его решение

```python

x=np.linspace(-15,15,201)

      
def f(x):
    return (np.exp(x)+2*x-x**3-1)

plt.plot(x, f(x))

plt.title("График f(x)")
plt.xlabel("Значения x")
plt.ylabel("Значения f(x)")
plt.xlim(-3,5)
plt.ylim(-5,20)
plt.legend(labels=["f(x)"])

plt.grid(True)
plt.show()
x1=fsolve(f, -2)
print(x1,x1**2-1)
x2=fsolve(f, 0)
print(x2,x2**2-1)
x3=fsolve(f, 2.5)
print(x3,x3**2-1)
x4=fsolve(f, 4)
print(x4,x4**2-1)
```

Решите систему уравнений и неравенств:
y = x2 – 1


exp(x) + x∙(1 – y) > 1

Эта система равносильна неравенству exp(x)+2x-x^3-1>0
решением будут все участки графика, которые выше оси Х.
